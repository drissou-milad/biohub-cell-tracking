```python
from src import config
from src.dataset import BioHubDataset

DATASET = BioHubDataset(config.DATASET_PATH)

print(len(DATASET.train_samples))
print(len(DATASET.test_samples))
```

```python
sample = DATASET.train_samples[0]

print(sample)
```

```python
sample_name = sample

print(sample_name)
```

```python
volume = DATASET.load_volume(sample_name)

print(volume.shape)
print(volume.dtype)
```

```python
import matplotlib.pyplot as plt

t = 0
z = 32

plt.figure(figsize=(6,6))
plt.imshow(volume[t, z], cmap="gray")
plt.title(f"Time={t}, Z={z}")
plt.axis("off")
plt.show()
```

```python
print("Minimum:", volume[:].min())
print("Maximum:", volume[:].max())
print("Mean:", volume[:].mean())
```

```python
fig, axes = plt.subplots(1, 5, figsize=(18,4))

for i, t in enumerate([0, 20, 40, 60, 80]):
    axes[i].imshow(volume[t, 32], cmap="gray")
    axes[i].set_title(f"t={t}")
    axes[i].axis("off")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(12,6))

zs = [0, 8, 16, 24, 32, 40, 48, 56]

for ax, z in zip(axes.flat, zs):
    ax.imshow(volume[0, z], cmap="gray")
    ax.set_title(f"z={z}")
    ax.axis("off")

plt.tight_layout()
```
